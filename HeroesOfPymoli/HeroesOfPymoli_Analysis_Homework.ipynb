```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish bins for ages
age_bins=[0,9.9,14.90,19.90,24.90,29.90,34.90,39.90,9999]
Groups= ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]


#Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_data["Age Group"]= pd.cut(purchase_data["Age"],age_bins, labels=Groups)

purchase_data


#Calculate the numbers and percentages by age group
Age_grouped = purchase_data.groupby("Age Group")

total_counts_of_age = Age_grouped["SN"].nunique()

age_percentage = (total_counts_of_age/players_total)*100



#Create a summary data frame to hold the results
Age_demograph = pd.DataFrame({"Percentage of Players":age_percentage, "Total Counts": total_counts_of_age})


#Optional: round the percentage column to two decimal points
#??


#Display Age Demographics Table

Age_demograph
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Run basic calculations to obtain the results in the table below
Big_Spenders = purchase_data.groupby("SN")
Purchase_count_by_Spender = Big_Spenders["Purchase ID"].count()
Average_spender_price = Big_Spenders["Price"].mean()
Total_spender_purchase = Big_Spenders["Price"].sum()




#Create a summary data frame to hold the results
Top_spenders = pd.DataFrame({"Purchase Count":Purchase_count_by_Spender,
                                 "Average Purchase Price":Average_spender_price,
                                 "Total Purchase Value":Total_spender_purchase,
                                 })

#Sort the total purchase value column in descending order
Sorted_spenders = Top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()

#Optional: give the displayed data cleaner formatting
#??

#Display a preview of the summary data frame
Sorted_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Items = purchase_data[["Item ID","Item Name","Price"]]

Item_group = Items.groupby(["Item ID","Item Name"])


Purchase_count_Items = Item_group["Price"].count()

Total_value_purchase = (Item_group["Price"].sum())
                                         
Item_price_value = Total_value_purchase/Purchase_count_Items                                     

Popular_items = pd.DataFrame({"Purchase Count":Purchase_count_Items,
                                 "Item Price":Item_price_value,
                                 "Total Purchase Value":Total_value_purchase,
                                 })
                                         
Popular_in_order =  Popular_items.sort_values(["Purchase Count"], ascending=False).head()
                                         
Popular_in_order                                         
```
